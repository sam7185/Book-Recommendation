```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")
print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv(f"{path}/books.csv")
df
```

```python
print(f"Shape of data: {df.shape}")
print("######################################")
print(f"dataframe type: ")
print(df.dtypes)
print("######################################")
for column in df.columns:
    unique_values = df[column].nunique()
    print(f"Unique values in column '{column}': {unique_values}")
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(),cbar=False, ax=ax)
plt.xlabel("missing values")
plt.ylabel("data columns")
```

```python
df["missing_description"] = np.where(df["description"].isnull(),1,0)
df["age_of_book"] = datetime.now().year - df["published_year"]
```

```python
column_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = df[column_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm")
plt.title("Correlation heatmap")
plt.plot()
```

```python
data = df[(~df["description"].isna() & ~df["num_pages"].isna() & ~df["average_rating"].isna() & ~df["published_year"].isna())]
data.shape
```

```python
data["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
data["words_in_description"] = data["description"].str.split().str.len()
data.head(10)
```

```python
data_with_good_description = data[data["words_in_description"] >= 25]
data_with_good_description.shape
```

```python
data_with_good_description["title_&_subtitle"] = (
    np.where(data_with_good_description["subtitle"].isna(),data_with_good_description["title"],
        data_with_good_description[["title","subtitle"]].astype(str).agg(" ".join, axis=1))
)
```

```python
data_with_good_description
```

```python
data_with_good_description["tagged_description"] = data_with_good_description[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
data_with_good_description.drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1).to_csv("books_dataset.csv",index=False)
```
